```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Stroke Prediction Dataset
stroke_df = pd.read_csv('encoded_stroke_data.csv')

# Determine the number of intervals for average glucose level
gluco_min = stroke_df['avg_glucose_level'].min()  # minimum glucose level
gluco_max = stroke_df['avg_glucose_level'].max()  # maximum glucose level
interval_size = 20  # set the size of each interval

# Determine the intervals
intervals = [(i, i+interval_size) for i in range(int(gluco_min), int(gluco_max)+1, interval_size)]

# Print the intervals and number of observations
for interval in intervals:
    obs = ((stroke_df['avg_glucose_level'] >= interval[0]) & (stroke_df['avg_glucose_level'] < interval[1])).sum()
    print(f'{interval}: {obs}')

# Create a histogram of average glucose level with intervals and a fade effect
hist, bin_edges = np.histogram(stroke_df['avg_glucose_level'], bins=[interval[0] for interval in intervals]+[intervals[-1][1]])
colors = ['#EEEEFF', '#7f7caf', '#9fb4c7', '#28587b', '#223843']
for i, interval in enumerate(intervals):
    plt.hist(stroke_df['avg_glucose_level'], bins=[interval[0], interval[1]], alpha=1-(i/len(intervals)), color='#7f7caf', edgecolor='black')
plt.xlabel('Nível médio de glicose')
plt.ylabel('Frequência')
plt.title('Histograma do nível médio de glicose')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the Stroke Prediction Dataset
stroke_df = pd.read_csv('encoded_stroke_data.csv')

# Determine the number of intervals for BMI
bmi_min = stroke_df['bmi'].min()  # minimum BMI
bmi_max = stroke_df['bmi'].max()  # maximum BMI
interval_size = 10  # set the size of each interval

# Determine the first class interval starting at the minimum BMI and ending at the first multiple of interval size
first_interval_size = interval_size - (bmi_min % interval_size)
if first_interval_size == interval_size:
    first_interval_size = 0
n_intervals = int((bmi_max - bmi_min - first_interval_size) / interval_size) + 1  # calculate the number of intervals
intervals = [(bmi_min, bmi_min + first_interval_size)] + [(bmi_min + first_interval_size + i*interval_size, bmi_min + first_interval_size + (i+1)*interval_size) for i in range(n_intervals-1)]

# Print the intervals with the number of observations
for interval in intervals:
    interval_df = stroke_df[(stroke_df['bmi'] >= interval[0]) & (stroke_df['bmi'] < interval[1])]
    count = len(interval_df)
    print(f"{interval[0]:.1f} - {interval[1]:.1f}: {count} observations")

# Create a histogram of BMI with intervals and a fade effect
hist, bin_edges = np.histogram(stroke_df['bmi'], bins=[interval[0] for interval in intervals]+[intervals[-1][1]])
colors = ['#EEEEFF', '#7f7caf', '#9fb4c7', '#28587b', '#223843']
for i, interval in enumerate(intervals):
    plt.hist(stroke_df['bmi'], bins=[interval[0], interval[1]], alpha=1-(i/n_intervals), color='#7f7caf', edgecolor='black')
plt.xlabel('IMC')
plt.ylabel('Frequência')
plt.title('Histograma do IMC')
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Load the Stroke Prediction Dataset
stroke_df = pd.read_csv('healthcare-dataset-stroke-data.csv')

# Create a cross-tabulation of ever_married and work_type
ct = pd.crosstab(stroke_df['ever_married'], stroke_df['work_type'])

# Calculate Cramer's V
import scipy.stats as stats

# Perform the chi-square test of independence
chi2_stat, p_val, dof, expected = stats.chi2_contingency(ct)

# Print the results
print(f"Chi-square statistic: {chi2_stat:.2f}")
print(f"p-value: {p_val:.10f}")

```

```python
import pandas as pd
import numpy as np

# carregar o dataset
stroke_df = pd.read_csv('cleaned_stroke_data.csv')

# separar os dados da variável idade em grupos, de acordo com ever_married
married = stroke_df[stroke_df['ever_married'] == 'Yes']['age']
not_married = stroke_df[stroke_df['ever_married'] == 'No']['age']

# calcular as medidas-resumo em cada grupo
married_mean = np.mean(married)
not_married_mean = np.mean(not_married)
married_std = np.std(married)
not_married_std = np.std(not_married)

# calcular a variância em cada grupo e ponderar pela quantidade de observações
married_var = np.var(married, ddof=1) * len(married) / len(stroke_df)
not_married_var = np.var(not_married, ddof=1) * len(not_married) / len(stroke_df)

# calcular a média das variâncias ponderadas
weighted_var = married_var + not_married_var

# comparar com a variância total da variável idade
total_var = np.var(stroke_df['age'], ddof=1)
if weighted_var <= total_var:
    print("A variância ponderada é menor ou igual à variância total.")
else:
    print("A variância ponderada é maior do que a variância total.")

# calcular o coeficiente de correlação de Pearson
correlation = stroke_df['age'].corr(stroke_df['ever_married'].astype('category').cat.codes)
print(f"O coeficiente de correlação de Pearson é {correlation:.2f}.")
```
